Load and inspect the data


```python
import pandas as pd

df = pd.read_excel('/content/Dataset for Data Analytics.xlsx')
print(df.shape)
print(df.head())
print(df.dtypes)
print(df.info())
```

Identify missing/null values

```python
print(df.isnull().sum())
```

Decide how to handle the missing values

```python
df['CouponCode'] = df['CouponCode'].fillna('No Coupon')
print(df['CouponCode'].value_counts())
```

Check for duplicates

```python
print("Full row duplicates:", df.duplicated().sum())
print("Duplicate OrderIDs:", df['OrderID'].duplicated().sum())
```

Fix data formats

```python
# Date format
df['Date'] = pd.to_datetime(df['Date'])
print(df['Date'].min(), df['Date'].max())

# Text columns - check for inconsistent casing/whitespace
for col in ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource', 'ShippingAddress']:
    print(col, df[col].unique()[:10])

# Strip whitespace and standardize case if needed
for col in ['Product', 'PaymentMethod', 'OrderStatus', 'ReferralSource']:
    df[col] = df[col].str.strip()
```

Validate numeric fields and round to 2 decimals

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)

# Sanity check: does TotalPrice = Quantity * UnitPrice?
df['Calculated_Total'] = (df['Quantity'] * df['UnitPrice']).round(2)
mismatches = df[df['Calculated_Total'] != df['TotalPrice']]
print("Mismatches:", len(mismatches))
```

Outlier scan (light pass, full analysis happens in Project 2)

```python
print(df[['Quantity', 'UnitPrice', 'TotalPrice', 'ItemsInCart']].describe())
```
