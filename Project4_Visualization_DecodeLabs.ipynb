```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Cleaned_Dataset_Project1.csv')
df['Date'] = pd.to_datetime(df['Date'])
```

Chart 1: Revenue by Product (horizontal bar, sorted, with action title)

```python
revenue_by_product = df.groupby('Product')['TotalPrice'].sum().sort_values()

fig, ax = plt.subplots(figsize=(8,5))
bars = ax.barh(revenue_by_product.index, revenue_by_product.values, color='#4472C4')

# Highlight the top one
bars[-1].set_color('#1F4E96')

ax.set_title('Chairs and Printers Drive ~31% of Total Revenue Combined', fontsize=13, fontweight='bold', loc='left')
ax.set_xlabel('Total Revenue ($)')

# Direct labeling
for bar in bars:
    width = bar.get_width()
    ax.text(width + 2000, bar.get_y() + bar.get_height()/2, f'${width:,.0f}', va='center')

plt.tight_layout()
plt.show()
```

Chart 2: Revenue by Referral Source

```python
revenue_by_ref = df.groupby('ReferralSource')['TotalPrice'].sum().sort_values()

fig, ax = plt.subplots(figsize=(8,5))
bars = ax.barh(revenue_by_ref.index, revenue_by_ref.values, color='#A6A6A6')
bars[-1].set_color('#1F4E96')

ax.set_title('Instagram is the Top Revenue-Generating Referral Channel', fontsize=13, fontweight='bold', loc='left')
ax.set_xlabel('Total Revenue ($)')

for bar in bars:
    width = bar.get_width()
    ax.text(width + 2000, bar.get_y() + bar.get_height()/2, f'${width:,.0f}', va='center')

plt.tight_layout()
plt.show()
```

Chart 3: Monthly revenue trend (line chart)

```python
monthly = df.groupby(df['Date'].dt.to_period('M'))['TotalPrice'].sum()

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(monthly.index.astype(str), monthly.values, color='#1F4E96', marker='o')
ax.set_title('Monthly Revenue Trend', fontsize=13, fontweight='bold', loc='left')
ax.set_ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Executive Summary (SCR framework)

### ***Situation: The business operates across seven product categories and five customer referral channels, generating a total of $1,265,162 in revenue from 1,200 orders.***


### ***Complication: Revenue is concentrated in a few areas — Chairs and Printers together account for nearly 31% of total revenue, while Phones lag behind at just 12%. On the acquisition side, Instagram and Email drive over 42% of revenue between them, while Referral traffic contributes the least.***


### ***Resolution: Marketing and inventory budgets should be reallocated to prioritize Chairs and Printers, and acquisition spend should shift further toward Instagram and Email campaigns, while the underperforming Referral channel and Phone category should be reviewed for repositioning or reduced investment.***
